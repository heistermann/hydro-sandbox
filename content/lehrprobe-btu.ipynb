## Das ABC der Abflussmodellierung

**Parameter des abc-Modells**
- `a`: Anteil des Niederschlag, der ins GW infiltriert
- `b`: Anteil des Niederschlags, der verdunstet
- `c`: Anteil des GW-Speichers, der zum Abfluss gelangt

Es muss stets gelt: `a` + `b` <= 1 und `c <= 1`.

```python
# Beispiel fÃ¼r eine Niederschlagszeitreihe
P = np.repeat(0,50)
P[[10,12,40]] = 100
```

```python
plot_abc_model(P)
```
